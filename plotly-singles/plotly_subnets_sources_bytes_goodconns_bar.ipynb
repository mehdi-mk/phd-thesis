```python
import plotly.graph_objects as go
import pandas as pd
import math

numrows = 256

df_original = pd.read_csv('/Volumes/GoogleDrive/My Drive/Thesis/Plotly/subnets_outgoing_sources_goodconns.csv')

max_bytes = math.log10(df_original["TotalBytes"].max())
min_bytes = math.log10(df_original["TotalBytes"].min())

df = df_original.drop(df_original[df_original.Week != 'Feb 2020'].index)

# print(df)
totalTraffic = df["TotalBytes"].sum()

x_tics = []
for i in range(numrows): x_tics.append(i)

fiftyPerc = []
for i in range(numrows): fiftyPerc.append(0.5 * totalTraffic) 
    
fivePerc = []
for i in range(numrows): fivePerc.append(0.05 * totalTraffic)
    
onePerc = []
for i in range(numrows): onePerc.append(0.01 * totalTraffic)

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = fiftyPerc,
        name = "50% Total",
        line = dict(color = 'red', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = fivePerc,
        name = "5% Total",
        line = dict(color = 'orange', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = onePerc,
        name = "1% Total", 
        line = dict(color = 'green', dash = 'dot')
    ))

fig.add_trace(
    go.Bar(
        x = df["Rank"],
        y = df["TotalBytes"],
        showlegend = False       
    ))
fig.update_traces(marker_color = 'rgb(8,48,107)')

# x_vals = df["SourceSubnet"].tolist()

fig.update_layout(font_size=20, width=1600, xaxis_range=[0,256], yaxis_range=[min_bytes, max_bytes])
fig.update_xaxes(title = "Subnet Ranks" )
fig.update_yaxes(title="Total Bytes", type="log", 
                 tickvals=[1048576, 10485760, 104857600, 1073741824, 10737418240, 107374182400, 
                           1099511627776, 10995116277760, 109951162777600],
                 ticktext=["1 MB", "10 MB", "100 MB", "1 GB", "10 GB", "100 GB", "1 TB", "10 TB", "100 TB"])


fig.show()
```

```python
fig.write_image('/Volumes/GoogleDrive/My Drive/Thesis/Plotly/Subnets_Out_TotalBytes_Feb2020.png')
```
