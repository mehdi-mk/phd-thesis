```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

numrows = 256 + 1

df = pd.read_csv('/Volumes/GoogleDrive/My Drive/Thesis/Plotly/subnets_outgoing_sources_2-8Feb2020.csv', nrows = numrows)
    
totalInbound = 0
for i in range(numrows): totalInbound += df["Inbound"][i]
    
totalOutbound = 0
for i in range(numrows): totalOutbound += df["Outbound"][i]
     
logInbound = []
for i in range(numrows): logInbound.append(np.log10(df["Inbound"][i]))
    
logOutbound = []
for i in range(numrows): logOutbound.append(np.log10(df["Outbound"][i]) * -1)

x_tics = []
for i in range(numrows): x_tics.append(i)
    
fiftyPercIn = []
for i in range(numrows): fiftyPercIn.append(np.log10(0.5 * totalInbound)) 
    
fivePercIn = []
for i in range(numrows ): fivePercIn.append(np.log10(0.05 * totalInbound))
    
onePercIn = []
for i in range(numrows): onePercIn.append(np.log10(0.01 * totalInbound))
    
fiftyPercOut = []
for i in range(numrows): fiftyPercOut.append(np.log10(0.5 * totalOutbound) * -1) 
    
fivePercOut = []
for i in range(numrows ): fivePercOut.append(np.log10(0.05 * totalOutbound) * -1)
    
onePercOut = []
for i in range(numrows): onePercOut.append(np.log10(0.01 * totalOutbound) * -1)

fig = go.Figure()

outbound_color = ['#969656',] * numrows
inbound_color = ['#567696',] * numrows

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = fiftyPercIn,
        name = "50%",
        line = dict(color = 'red', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = fivePercIn,
        name = "5%",
        line = dict(color = 'orange', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = onePercIn,
        name = "1%", 
        line = dict(color = 'green', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = fiftyPercOut,
        showlegend=False,
        line = dict(color = 'red', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = fivePercOut,
        showlegend=False,
        line = dict(color = 'orange', dash = 'dot')
    ))

fig.add_trace(
    go.Scatter(
        x = x_tics,
        y = onePercOut,
        showlegend=False,
        line = dict(color = 'green', dash = 'dot')
    ))

fig.add_trace(
    go.Bar(
        x = x_tics,
        y = logInbound,
        name = "Inbound",
        marker_color = inbound_color,
        offsetgroup=0,
    ))

fig.add_trace(
    go.Bar(
        x = x_tics,
        y = logOutbound,
        name = "Outbound",
        marker_color = outbound_color,
        offsetgroup=0,
    ))

#fig.update_traces(marker_color = 'rgb(8,48,107)')

x_vals = df["SourceSubnet"].tolist()

fig.update_xaxes( tickangle = 90, tickvals = x_tics, ticktext = x_vals, title = "Subnets" )
fig.update_yaxes(title = "Traffic", range=[-15, 15], tickvals = [-14.04, -11.03, -8.02, 0, 8.02, 11.03, 14.04],
                ticktext = ["-100 TB", "-100 GB", "-100 MB", "0", "100 MB", "100 GB", "100 TB"])
fig.update_layout(title = "Inbound and Outbound Traffic of Subnets (2-8 Feb 2020)")

fig.show()
```
