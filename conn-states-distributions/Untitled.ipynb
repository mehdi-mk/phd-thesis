```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# Generate some example data
dates = [datetime.datetime(2022, 1, 1) + datetime.timedelta(days=i*30) for i in range(12)]
values = [i for i in range(12)]

# Create the plot
fig, ax = plt.subplots()

# Plot the data
ax.plot(dates, values)

# Set the x-axis locator
locator = mdates.MonthLocator(interval=3)  # Show labels for every 3 months
ax.xaxis.set_major_locator(locator)

# Format the x-axis tick labels
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Rotate the tick labels for better readability (optional)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
